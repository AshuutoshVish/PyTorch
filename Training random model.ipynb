```python
#Working on Regression Model with random numbers
```

```python
import torch
from torch import nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
weight=0.7
bias=0.3
```

```python
start=0
stop=1
step=0.002
```

```python
X=torch.arange(start, stop, step).unsqueeze(dim=1)
y=weight * X + bias
```

```python
X[:10], y[:10]
```

```python
len(X), len(y)
```

```python
#Splitting the Random Data
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y,test_size=0.2)
```

```python
len(X_train), len(y_train), len(X_test), len(y_test), len(X), len(y)
```

```python
#Visualising the training data
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(X_train, y_train, c="r", s=4, label="Training data")
plt.legend(prop={"size": 7});
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(X_train, y_train, c="b", s=4, label="Training data")
plt.scatter(X_test, y_test, c="r", s=4, label="Testing data")
plt.legend(prop={"size": 5});
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight=nn.Parameter(torch.randn(1,
                                              requires_grad=True,
                                              dtype=torch.float))
        self.bias=nn.Parameter(torch.randn(1,
                                            requires_grad=True,
                                            dtype=torch.float))
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weight * x + self.bias
```

```python
torch.manual_seed(42)
model_LR = LinearRegressionModel()
list(model_LR.parameters())
```

```python
model_LR.state_dict()
```

```python
with torch.no_grad(): 
    y_pred = model_LR(X_test)
```

```python
y_pred = model_LR(X_test)
```
